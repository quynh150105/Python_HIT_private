Câu 5:
-, Khai báo và gán giá trị:
  <Tên_biến> = <Giá_trị>
  ví dụ: int a = 3
  Python sẽ tự định nghĩa kiểu định liệu cho biến.

-, Các toán tử trong python:
  +,Toán tử số học: + - * / %(chia lấy dư),**(lũy thừa), //(chia lấy phần nguyên)
  ví dụ:
  a = 10
  b = 3
  print(a + b)  
  print(a - b)  
  print(a * b)  
  print(a / b)
  print(a % b)
  print(a ** b)
  print(a // b)

  +, Toán tử logic: and, or, not
  ví dụ:
  print(a > 5 and b < 10)  # True
  print(a > 5 or b > 10)   # True
  print(not(a > 5))        # False

  +, Toán tử so sánh:
  ==, !=, >, <, >=, <=
  ví dụ:
  a = 5
  b = 3
  print(a == b)  # False
  print(a != b)  # True
  print(a > b)   # True
  print(a < b)   # False
  print(a >= b)  # True
  print(a <= b)  # False

  +, Toán tử gán: =, +=, -=, *=, /= %=
  ví dụ:
  c = 5
  c += 2  
  c *= 3  

-, Các kiểu dữ liệu trong python:
  +, Kiểu số (Number): int (số nguyên), float (số thực), complex (số phức)

  +, Kiểu chuỗi (String): Được bao bởi dấu nháy đơn ('') hoặc nháy kép ("")

  +, Kiểu danh sách (List): Danh sách các phần tử có thể thay đổi

  +, Kiểu bộ (Tuple): Danh sách các phần tử không thể thay đổi

  +, Kiểu từ điển (Dictionary): Cặp khóa-giá trị

  +, Kiểu tập hợp (Set): Tập hợp các phần tử không trùng lặp
